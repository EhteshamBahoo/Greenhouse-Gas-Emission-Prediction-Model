```python
df = pd.read_csv("Data/owid-co2-data.csv")

print("Dataset Info:")
print(df.info())

print("\nFirst Five Rows:")
display(df.head())

print("\nColumn Names:")
print(df.columns.to_list())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df = df[['year', 'country', 'co2', 'co2_per_capita', 'co2_per_gdp', 
         'methane', 'nitrous_oxide', 'primary_energy_consumption']]

df.fillna(method='ffill', inplace=True)

df['year'] = df['year'].astype(int)

print("\nDataset After Preprocessing:")
display(df.head())

print("\nRemaining Missing Values:")
print(df.isnull().sum())
```

```python
df.dropna(subset=['co2', 'co2_per_capita', 'co2_per_gdp', 
                  'methane', 'nitrous_oxide', 'primary_energy_consumption'], how='all', inplace=True)

df = df[df['year'] >= 1900]

df.reset_index(drop=True, inplace=True)

print("\nDataset After Removing NaN Rows and Filtering by Year:")
display(df.head())

print("\nRemaining Missing Values After Cleaning:")
print(df.isnull().sum())
```

```python
df['co2'].fillna(df['co2'].mean(), inplace=True)
df['co2_per_capita'].fillna(df['co2_per_capita'].mean(), inplace=True)
df['co2_per_gdp'].fillna(df['co2_per_gdp'].mean(), inplace=True)

df['primary_energy_consumption'].fillna(df['primary_energy_consumption'].median(), inplace=True)

print("\nFinal Missing Values Check:")
print(df.isnull().sum())
```

## ML model

```python
from sklearn.linear_model import LinearRegression
```

```python
features = ['primary_energy_consumption', 'methane', 'nitrous_oxide']
target = 'co2'
```

### Regression Model (Predict CO₂ Emissions)

```python
df_reg = df.dropna(subset=features + [target])
X_train, X_test, y_train, y_test = train_test_split(df_reg[features], df_reg[target], test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.5)
plt.xlabel("Actual CO₂ Emissions")
plt.ylabel("Predicted CO₂ Emissions")
plt.title("Actual vs. Predicted CO₂ Emissions")
plt.grid(True)
plt.show()
```

## Advanced Modeling

### Deep Learning with LSTM (Better CO₂ Forecasting)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
```

```python
co2_values = co2_timeseries.values.reshape(-1, 1)
```

```python
scaler = MinMaxScaler()
co2_scaled = scaler.fit_transform(co2_values)
```

```python
def create_sequences(data, seq_length=10):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 10
X, y = create_sequences(co2_scaled, seq_length)
```

```python
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(seq_length, 1)),
    LSTM(50),
    Dense(25, activation='relu'),
    Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))
```

```python
future_pred = model.predict(X_test[-1].reshape(1, 10, 1))  
future_pred = scaler.inverse_transform(future_pred)

print("\nPredicted Future CO₂ Levels:", future_pred)
```
